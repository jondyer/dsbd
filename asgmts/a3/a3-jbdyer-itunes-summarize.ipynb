## 5. Analyze the data

**5.i) Clean the data.**

```python
import pandas as pd
```

```python
data = pd.read_csv("3.csv")
data.head()
```

```python
data[data.app_name=='WhatsApp Messenger']
```

```python
# Some rows do not pull in star_rating and num_ratings 
# Drop these rows for the analysis
data = data.dropna()
data.describe(include='all')
```

```python
# Create a clean ratings column converts 189.5K to 189500 etc
# first create a function to apply to each value in the column
def clean_ratings(x):
    '''
    Takes a string x in the form "[fl][suffix] Ratings" where fl is some
    floating point or integer value, and suffix may be M (million), 
    K (thousand), or empty.
    
    Returns the converted value as an integer.
    '''
    first, _ = x.split()   # retrieve the [fl][suffix] part
    suffix = first[-1]
    
    val = 0
    if suffix.isalpha():    # check if there is a suffix
        if suffix=='K':        # thousands
            val = float(first[:-1]) * 1000
        elif suffix=='M':      # millions
            val = float(first[:-1]) * 1000000
    else:                   # otherwise it's all numeric
        val = float(first)
        
    return int(val)
```

**5.ii) Add new column**

```python
# now create the new column using that function
data['num_ratings_clean'] = data['num_ratings'].apply(clean_ratings)
data.head()
```

**5.iii) List top apps by rating and num_ratings**

```python
# first get just the columns we want
sub = data[['app_name', 'star_rating', 'num_ratings_clean']]
sub.head()
```

```python
# then sort by app name, followed by the attributes
sub.sort_values('app_name').sort_values(['star_rating', 'num_ratings_clean'], ascending=False).head(10)
```
